```python
from sklearn import preprocessing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Get dataset
df = pd.read_csv("./california_housing_train.csv", sep=",")
df
```

```python
plt.hist(df['total_bedrooms'], 50)
plt.show()
```

## Normalization

```python
# Normalize total_bedrooms column
x_array = np.array(df['total_bedrooms'])
total_bedrooms_normalized = preprocessing.normalize([x_array])
df['total_bedrooms_normalized'] = total_bedrooms_normalized[0]
```

```python
plt.hist(df['total_bedrooms_normalized'], 50)
plt.show()
```

## Standardization

```python
# Standardize total_bedrooms column
scaler = preprocessing.StandardScaler()
total_bedrooms_standardized = scaler.fit_transform(df[['total_bedrooms']])
df['total_bedrooms_standardized'] = total_bedrooms_standardized
```

```python
plt.hist(df['total_bedrooms_standardized'], 50)
plt.show()
```
