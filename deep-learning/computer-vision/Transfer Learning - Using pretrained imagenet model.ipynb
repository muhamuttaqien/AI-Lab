```python
import keras
import numpy as np
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications import vgg16, inception_v3, resnet50, mobilenet
from keras.applications.imagenet_utils import decode_predictions

import matplotlib.pyplot as plt
%matplotlib inline
```

## Load Pre-trained Model

```python
vgg_model = vgg16.VGG16(weights='imagenet')
```

```python
inception_model = inception_v3.InceptionV3(weights='imagenet')
```

```python
resnet_model = resnet50.ResNet50(weights='imagenet')
```

```python
mobilenet_model = mobilenet.MobileNet(weights='imagenet')
```

## Load Image

```python
image_path = './images/corn.jpeg' # cat, dog, horse, duck, bfly, tomato, corn, grape
original_image = load_img(image_path, target_size=(224, 224))
plt.imshow(original_image)
plt.show()
print('Original image size: ', original_image.size)
```

```python
# convert the image to a numpy array
numpy_image = img_to_array(original_image)
# plt.imshow(np.uint8(numpy_image))
# plt.show()
print('Numpy image size: ', numpy_image.shape)

# convert the image into batch format
batch_image = np.expand_dims(numpy_image, axis=0)
# plt.imshow(np.uint8(batch_image[0]))
# plt.show()
print('Batch image size: ', batch_image.shape)
```

Once we have the image in the right format, we can feed it to the network and get the predictions.

## Predict the Image

```python
# prepare the image for the VGG model
vgg_processed_image = vgg16.preprocess_input(batch_image.copy())
inception_processed_image = inception_v3.preprocess_input(batch_image.copy())
resnet_processed_image = resnet50.preprocess_input(batch_image.copy())
mobilenet_processed_image = mobilenet.preprocess_input(batch_image.copy())
```

```python
# get the predicted probabilities for each class
vgg_predictions = vgg_model.predict(vgg_processed_image)
inception_predictions = inception_model.predict(inception_processed_image)
resnet_predictions = resnet_model.predict(resnet_processed_image)
mobilenet_predictions = mobilenet_model.predict(mobilenet_processed_image)

# we will get top 5 predictions which is the default
predictions_list = {'vgg_predictions': vgg_predictions, 
                    'inception_predictions': inception_predictions, 
                    'resnet_predictions': resnet_predictions, 
                    'mobilenet_predictions': mobilenet_predictions}
```

```python
for key, value in predictions_list.items():
    predicted_label = decode_predictions(value)
    print('Prediction: ', key)
    print(predicted_label)
    print('---')
```
