## Build VOC Class

```python
class VOC(torch.utils.data.Dataset):
    
    """
    Pascal VOC datasets from http://host.robots.ox.ac.uk/pascal/VOC/voc2007/index.html
    """
    
    # training dataset root directories
    data_dir = 'JPEGImages'
    label_dir = 'SegmentationClass'
    
    # images extension
    data_extension = '.jpg'
    label_extension = '.png'
    
    num_classes = 12
    
    def __init__(self, 
                 root_dir, 
                 data_transform=None,  
                 label_transform=None):
        
        self.root_dir = root_dir
        self.data_transform = data_transform
        self.label_transform = label_transform
        
        # get the data and labels filepaths
        self.data = utils.get_files(os.path.join(root_dir, self.data_dir),
                                                 extension_filter=self.data_extension)
        
        self.labels = utils.get_files(os.path.join(root_dir, self.label_dir), 
                                                   extension_filter=self.label_extension)
        
    def __getitem__(self, index):
        
        data_path, label_path = self.data[index], self.labels[index]
        
        image, label = utils.pil_loader(data_path, label_path)
        
        if self.data_transform is not None:
            image = self.data_transform(image)
        
        if self.label_transform is not None:
            label = self.label_transform(label)
        
        # perform one-hot-encoding
        target = utils.one_hot_encode(self.num_classes, label)
        target = torch.FloatTensor(target)
        
        return image, label, target
    
    def __len__(self):
        return len(self.data)
```

## Sanity Check

```python
DATASETS_DIR = './data'
HEIGHT, WIDTH = 360, 480
BATCH_SIZE = 10
WORKERS = 4
```

```python
data_transform = transforms.Compose([transforms.Resize((HEIGHT, WIDTH)), 
                                     transforms.ToTensor()])

label_transform = transforms.Compose([transforms.Resize((HEIGHT, WIDTH), Image.NEAREST),
                                      transforms.ToTensor()])
```

```python
voc_set = VOC(DATASETS_DIR, data_transform=data_transform, label_transform=label_transform)

num_classes = voc_set.num_classes
```

```python
voc_loader = torch.utils.data.DataLoader(voc_set, batch_size=BATCH_SIZE, shuffle=True, num_workers=WORKERS)
```

```python
images, labels, targets = iter(voc_loader).next()
print("Images size:", images.size())
print("Labels size:", labels.size())
print("Targets size:", targets.size())
print("Number of class:", num_classes)
```
