```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## 3 Ways To Display Image

```python
image_path = './images/aurora.jpg'
```

```python
# PIL
image = Image.open(image_path)
image
```

```python
# CV2
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
# Matplotlib
image = mpimg.imread(image_path)
plt.imshow(image)
plt.show()
```

## Show Histograms

```python
plt.figure(figsize=(10,5))
for i, col in enumerate(['b', 'g', 'r']):
    hist = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])
    
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
plt.figure(figsize=(10,5))
plt.plot(hist, color='k')
plt.show()
```

## Histogram Equalization

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
eq_gray_image = cv2.equalizeHist(gray_image)
plt.imshow(eq_gray_image, cmap='gray')
plt.show()
```

```python
hist = cv2.calcHist([eq_gray_image], [0], None, [256], [0, 256])
plt.figure(figsize=(10,5))
plt.plot(hist, color='k')
plt.show()
```

```python
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
channels = cv2.split(image)
eq_channels = []
for ch in channels:
    eq_channels.append(cv2.equalizeHist(ch))
    
eq_image = cv2.merge(eq_channels)
plt.imshow(eq_image)
plt.show()
```

```python
plt.figure(figsize=(10,5))
for i, col in enumerate(['b', 'g', 'r']):
    hist = cv2.calcHist([eq_image], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])
    
plt.show()
```
