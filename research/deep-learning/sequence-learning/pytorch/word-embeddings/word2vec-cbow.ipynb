# Continuous Bag-of-Word Model

```python
import os
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize

import torch
import torch.nn as nn
from torch.autograd import Variable
```

```python
import warnings
warnings.filterwarnings('ignore')
```

## Set Helpers

```python
def text_to_data(text, context_window=2):
    
    data = []
    
    for i in range(context_window, len(text) - context_window):
        context = [
            text[i+e] for e in range(-context_window, context_window+1) if i+e != i
        ]
        target = text[i]
        
        data.append((context, target))
        
    return data
```

```python
def words_to_tensor(words: list, word2idx: dict, dtype=torch.FloatTensor):
    
    tensor = dtype([
        word2idx[word] for word in words
    ])
    
    tensor = tensor.to(device)
        
    return Variable(tensor)
```

```python
def get_prediction(context, model):
    
    model.eval()
    prediction = model(words_to_tensor(context, word2idx))
    _, index = torch.max(prediction, 1)
    
    return idx2word[index.item()]
```

```python
def check_accuracy(model):
    
    correct = 0
    for context, target in data:
        prediction = get_prediction(context, model)
        if prediction == target:
            correct += 1
            
    return correct/ len(data)
```

## Load Corpus

```python
corpus = """
As the eight strange beings applauded, one of them even cupping a hand over her lipsticked mouth to cheer, 
Joel tried to grasp what was happening. The nine of them sat in a fire rimmed cavern around a conference 
table shaped from warm volcanic rock. A chandelier of human bones dangled from the cavern’s ceiling, and it 
rattled around at random like wind chimes. A massive goat-man with reddish-black skin and wicked horns on his 
head towered above the seven others, who flanked him to either side. They looked like pure stereotype. A fat slob 
with sixteen chins, a used car saleman looking guy with gold and silver jewelry all over him, a sultry dominatrix 
in skin tight leather. On the other side a disheveled looking college drop out, a pretty boy staring in a mirror, 
a bald, muscular dude who looked like someone’s pissed off step-dad and a sour faced woman glancing jealously 
around the room. Just where the hell was he? Joel concentrated on his last memory. He remembered highlighting 
pages as his private jet, “The Holy Gust,” flew over the sapphire waters of the Bahamas. He had been reviewing 
his sermon for Sunday – dotting the I’s and crossing the crosses, a little god humor there, praise him – and the 
pilot’s voice had crackled over the intercom about turbulence. Kimberly, his personal assistant, had taken his 
plow out of her mouth and put on her seat belt. The plane had shook and then"""
```

## Pre-process Corpus

```python
words = word_tokenize(corpus.lower())
```

```python
vocab = set(words)
word2idx = {word:idx for idx,word in enumerate(vocab)}
idx2word = {idx:word for idx,word in enumerate(vocab)}
```

```python
data = text_to_data(words, 2)
```

## Set Configs

```python
LR = 1e-3
N_EPOCHS = 500

vocab_size = len(vocab)
EMBEDDING_DIM = 100
```

```python
is_cuda = torch.cuda.is_available()

if is_cuda: device = torch.device('cuda')
else: device = torch.device('cpu')
```

## Build [CBOW](https://arxiv.org/pdf/1301.3781.pdf) Network

```python
class CBOW(nn.Module):
    
    def __init__(self, input_size, embedding_dim):
        super().__init__()
        
        self.vocab_size = input_size
        
        self.embedding_layer = nn.Embedding(self.vocab_size, embedding_dim)
        self.fc_layer = nn.Linear(embedding_dim, self.vocab_size)
        self.softmax = nn.LogSoftmax()
        
    def forward(self, x):
        
        x = x.long()
        embeds = self.embedding_layer(x)
        embeds = torch.mean(embeds, dim=0).view(1,-1)
        
        output = self.fc_layer(embeds)
        output = self.softmax(output)
        
        return output
```

#### Initialize CBOW network

```python
cbow = CBOW(vocab_size, EMBEDDING_DIM)
cbow.to(device)
```

## Set Loss Function

```python
nll_loss = torch.nn.NLLLoss()
```

## Set Optimizer

```python
optimizer = torch.optim.SGD(cbow.parameters(), lr=LR)
```

## Train CBOW Network

```python
print_every = 100
train_losses = []

cbow.train()
for epoch in range(N_EPOCHS):
    
    train_loss = 0
    
    for context, target in data:
        
        context = words_to_tensor(context, word2idx)
        target = words_to_tensor([target], word2idx, dtype=torch.LongTensor)
        
        optimizer.zero_grad()
        output = cbow(context)
        loss = nll_loss(output, target)
        
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()
        
    if epoch%print_every == 0:
        accuracy = check_accuracy(cbow)
        print('Epoch: {}/{}, '.format(epoch, N_EPOCHS),
              'Accuracy: {:.6f}'.format(accuracy))
        
    train_losses.append(train_loss)
```

```python
plt.figure(figsize=(10,5))
plt.title("Train Loss of CBOW Network")
plt.plot(train_losses, label="Train Loss", color="green")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

if not os.path.exists('./images/'): os.makedirs('./images/')
plt.savefig('./images/cbow_final_train_loss.png')
plt.show()
```

```python
plt.figure(figsize=(20,20))
plt.scatter(embeds_pca[:, 0], embeds_pca[:, 1])
words = list(vocab)
for i, word in enumerate(words):
    plt.annotate(word, xy=(embeds_pca[i, 0], embeds_pca[i, 1]))

if not os.path.exists('./images/'): os.makedirs('./images/')
plt.savefig('./images/cbow_pca_word_vectors.png')
plt.show()
```
