```python
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image

plt.style.use('ggplot')
```

## Continous Distribution

#### 1) Uniform Distribution

```python
# import uniform distribution
from scipy.stats import uniform
```

```python
# random numbers from uniform distribution
n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc=start, scale=width)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.distplot(data_uniform,
                  bins=100,
                  kde=True,
                  color='green')

ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')

plt.show()
```

#### 2) Normal Distribution

```python
# import normal distribution
from scipy.stats import norm
```

```python
# random numbers from normal distribution
n = 10000
start = 0
width = 1
data_normal = norm.rvs(size=n, loc=start, scale=width)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.distplot(data_normal,
                  bins=100,
                  kde=True,
                  color='green')

ax.set(xlabel='Normal Distribution', ylabel='Frequency')

plt.show()
```

## Discrete Distribution

#### 1) Poisson Distribution

```python
# import poisson distribution
from scipy.stats import poisson
```

```python
# random numbers from poisson distribution
n = 10000
event_rate = 3
data_poisson = poisson.rvs(size=n, mu=event_rate)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.distplot(data_poisson,
                  bins=100,
                  kde=True,
                  color='blue')

ax.set(xlabel='Poisson Distribution', ylabel='Frequency')

plt.show()
```

#### 2) Binomial Distribution

```python
# import binomial distribution
from scipy.stats import binom
```

```python
# random numbers from binomial distribution
n = 10000
num_trials = 10
success_prob = 0.8
data_binomial = binom.rvs(size=n, n=num_trials, p=success_prob)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.distplot(data_binomial,
                  bins=100,
                  kde=True,
                  color='blue')

ax.set(xlabel='Binomial Distribution', ylabel='Frequency')

plt.show()
```

#### 3) Bernoulli Distribution

```python
# import bernoulli bernoulli
from scipy.stats import bernoulli
```

```python
# random numbers from bernoulli distribution
n = 10000
success_prob = 0.6
data_bernoulli = bernoulli.rvs(size=n, p=success_prob)
```

```python
plt.figure(figsize=(10, 5))
ax = sns.distplot(data_bernoulli,
                  bins=100,
                  kde=True,
                  color='blue')

ax.set(xlabel='Bernoulli Distribution', ylabel='Frequency')

plt.show()
```
